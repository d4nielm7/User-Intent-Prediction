### Added Rating Count (RANDOM)

```python
rating_count_from_history = TG_exploded.groupby('ProdID_List').size().reset_index(name='RatingCount')
```

```python
item_data_updated = pd.merge(df, rating_count_from_history, left_on='ProdID', right_on='ProdID_List', how='left')

item_data_updated['RatingCount'] = item_data_updated.apply(
    lambda row: row['RatingCount'] if pd.notna(row['RatingCount']) else row['ReviewCount'] + np.random.randint(1, 50),
    axis=1
)

upper_limit_multiplier = 2

min_rating_threshold = 5

item_data_updated['RatingCount'] = item_data_updated.apply(
    lambda row: row['RatingCount'] if row['RatingCount'] > row['ReviewCount'] else 
    np.random.randint(max(row['ReviewCount'] + 1, min_rating_threshold), max(int(row['ReviewCount'] * upper_limit_multiplier), min_rating_threshold + 1)),
    axis=1
)

column_order = ['ID','ProdID', 'TimeStamp', 'Rating', 'RatingCount', 'ReviewCount', 'Category', 'Brand', 'Name', 'ImageURL', 'Description', 'Price', 'Tags']

# Apply the new column order to the dataframe
df = item_data_updated[column_order]
```

```python
df.head(10)
```

### Tag Extraction and Text Cleaning

```python
tqdm.pandas()

nlp = spacy.load("en_core_web_sm")

def clean_and_extract_tags(text):
    doc = nlp(text.lower())
    tags = [token.text for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]
    return ', '.join(tags)


columns_to_extract_tags_from = ['Category', 'Brand', 'Description', 'Tags']

for column in columns_to_extract_tags_from:
    df[column] = df[column].progress_apply(clean_and_extract_tags)
```

```python
df.head()
```

## Feature Engineering

### Recency

```python
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'], utc=True, errors='coerce')

df['TimeStampDay'] = df['TimeStamp'].dt.date

df = df.dropna(subset=['TimeStampDay'])

customer_data = df.groupby('ID')['TimeStampDay'].max().reset_index()

most_recent_date =df['TimeStampDay'].max()

customer_data['TimeStampDay'] = pd.to_datetime(customer_data['TimeStampDay'])

most_recent_date = pd.to_datetime(most_recent_date)

customer_data['Days_Since_Last_Purchase'] = (most_recent_date - customer_data['TimeStampDay']).dt.days

customer_data.drop(columns=['TimeStampDay'], inplace=True)

customer_data.head()
```

### Frequency

```python
total_transactions = df.groupby('ID')['TimeStamp'].nunique().reset_index()
total_transactions.rename(columns={'TimeStamp': 'Total_Transactions_Unique'}, inplace=True)

total_products_purchased = df.groupby('ID').size().reset_index(name='Total_Products_Purchased')

customer_data = pd.merge(customer_data, total_transactions, on='ID')
customer_data = pd.merge(customer_data, total_products_purchased, on='ID')

customer_data.head()
```

### Monetary

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

df = df.dropna(subset=['Price'])  

df['Total_Spend'] = df['Price']  
total_spend = df.groupby('ID')['Total_Spend'].sum().reset_index()

average_transaction_value = total_spend.merge(total_transactions, on='ID')
average_transaction_value['Average_Transaction_Value'] = average_transaction_value['Total_Spend'] / average_transaction_value['Total_Transactions_Unique']

customer_data = pd.merge(customer_data, total_spend, on='ID')
customer_data = pd.merge(customer_data, average_transaction_value[['ID', 'Average_Transaction_Value']], on='ID')

customer_data.head()
```

### Product Diversity

```python
unique_products_purchased = df.groupby('ID')['ProdID'].nunique().reset_index()
unique_products_purchased.rename(columns={'ProdID': 'Unique_Products_Purchased'}, inplace=True)

customer_data = pd.merge(customer_data, unique_products_purchased, on='ID')

customer_data.head()
```

### Behavioral Features

```python
df = df.sort_values(['ID', 'TimeStamp'])

df['Day_Of_Week'] = df['TimeStamp'].dt.dayofweek
df['Hour'] = df['TimeStamp'].dt.hour

days_between_purchases =df.groupby('ID')['TimeStampDay'].apply(lambda x: (x.diff().dropna()).apply(lambda y: y.days))

average_days_between_purchases = days_between_purchases.groupby('ID').mean().reset_index()
average_days_between_purchases['TimeStampDay'] = average_days_between_purchases['TimeStampDay'].round().astype(int)

average_days_between_purchases.rename(columns={'TimeStampDay': 'Average_Days_Between_Purchases'}, inplace=True)

favorite_shopping_day = df.groupby(['ID', 'Day_Of_Week']).size().reset_index(name='Count')
favorite_shopping_day = favorite_shopping_day.loc[favorite_shopping_day.groupby('ID')['Count'].idxmax()][['ID', 'Day_Of_Week']]

favorite_shopping_hour = df.groupby(['ID', 'Hour']).size().reset_index(name='Count')
favorite_shopping_hour = favorite_shopping_hour.loc[favorite_shopping_hour.groupby('ID')['Count'].idxmax()][['ID', 'Hour']]

customer_data = pd.merge(customer_data, average_days_between_purchases, on='ID', how='left', suffixes=('', '_x'))
customer_data = pd.merge(customer_data, favorite_shopping_day, on='ID', how='left', suffixes=('', '_x'))
customer_data = pd.merge(customer_data, favorite_shopping_hour, on='ID', how='left', suffixes=('', '_x'))

customer_data = customer_data.loc[:,~customer_data.columns.str.endswith('_x')]

customer_data.head(20)
```

### Seasonality and Trends

```python
df['Year'] = df['TimeStamp'].dt.year
df['Month'] = df['TimeStamp'].dt.month

monthly_spending = df.groupby(['ID', 'Year', 'Month'])['Total_Spend'].sum().reset_index()

seasonal_buying_patterns = monthly_spending.groupby('ID')['Total_Spend'].agg(['mean', 'std']).reset_index()
seasonal_buying_patterns.rename(columns={'mean': 'Monthly_Spending_Mean', 'std': 'Monthly_Spending_Std'}, inplace=True)

seasonal_buying_patterns['Monthly_Spending_Std'].fillna(0, inplace=True)

def calculate_trend(spend_data):
    if len(spend_data) > 1:
        x = np.arange(len(spend_data))
        slope, _, _, _, _ = linregress(x, spend_data)
        return slope
    else:
        return 0

spending_trends = monthly_spending.groupby('ID')['Total_Spend'].apply(calculate_trend).reset_index()
spending_trends.rename(columns={'Total_Spend': 'Spending_Trend'}, inplace=True)

customer_data = pd.merge(customer_data, seasonal_buying_patterns, on='ID')
customer_data = pd.merge(customer_data, spending_trends, on='ID')

customer_data.head(20)
```
