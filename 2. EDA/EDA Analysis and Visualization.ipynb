## Item Analysis

### Top Brands

```python
def visualize_top_brands(data):
    top_brands = data['Brand'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_brands.values, y=top_brands.index, palette='viridis', legend=False)
    plt.title('Top 10 Brands by Count')
    plt.xlabel('Count')
    plt.ylabel('Brand')
    plt.show()

visualize_top_brands(Item_data)
```

### Top Categories

```python
def visualize_top_categories(data):
    top_categories = data['Category'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_categories.values, y=top_categories.index, palette='magma')
    plt.title('Top 10 Categories by Count')
    plt.xlabel('Count')
    plt.ylabel('Category')
    plt.show()
visualize_top_categories(Item_data)
```

```python
bins = [0, 5, 10, 20, 50, 100, 200, 400, 800]
labels = ['Very Cheap', 'Cheap', 'Somewhat Cheap', 'Normal Pricing', 'Moderately Pricey', 'Expensive', 'Very Expensive', 'Premium']

# Binning the data based on 'Price' column (assuming 'Price' is the correct column in your dataset)
Item_data['price_bin'] = pd.cut(Item_data['Price'], bins=bins, labels=labels)

# Creating a frequency table for the binned prices
frequency_table = Item_data['price_bin'].value_counts().sort_index()

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Plot for price distribution
ax.bar(frequency_table.index.astype(str), frequency_table.values, color='blue')
ax.set_title('Frequency Price Distribution')
ax.set_xlabel('Price Bins')
ax.set_ylabel('Number of Products')
ax.tick_params(axis='x', rotation=45)

# Add labels to the bars
ax.bar_label(ax.containers[0])

plt.tight_layout()
plt.show()
```

### Analyze User based on Spending

```python
Item_data['TimeStamp'] = pd.to_datetime(Item_data['TimeStamp'], errors='coerce')

monthly_spend = Item_data.groupby(Item_data['TimeStamp'].dt.to_period('M')).agg({'Total_Spend': 'sum'}).reset_index()

monthly_spend['TimeStamp'] = monthly_spend['TimeStamp'].dt.to_timestamp()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_spend, x='TimeStamp', y='Total_Spend')
plt.ylim(0)  
plt.title('Monthly Spending Trend')
plt.xlabel('Month')
plt.ylabel('Total Spend')
plt.show()
```

```python
merged_data = pd.merge(purchase_history, Item_data[['ProdID', 'price_bin']], left_on='ProdID_List', right_on='ProdID', how='left')
```

```python
price_bin_mapping = {
    'Very Cheap': 1,
    'Cheap': 2,
    'Somewhat Cheap': 3,
    'Normal Pricing': 4,
    'Moderately Pricey': 5,
    'Expensive': 6,
    'Very Expensive': 7,
    'Premium': 8
}
```

```python
merged_data['price_bin_value'] = merged_data['price_bin'].map(price_bin_mapping)
merged_data['price_bin_value'] = pd.to_numeric(merged_data['price_bin_value'], errors='coerce')
```

```python
customer_spending = merged_data.groupby('ID')['price_bin_value'].mean().reset_index()
customer_spending.columns = ['ID', 'Average_Spending_Score']
def categorize_spender(score):
    if score <= 2.5:
        return 'Very Low Spender'
    elif score <= 3.5:
        return 'Low Spender'
    elif score <= 4.5:
        return 'Moderate Spender'
    elif score <= 5.5:
        return 'Average Spender'
    elif score <= 6.5:
        return 'High Spender'
    elif score <= 7.5:
        return 'Very High Spender'
    else:
        return 'Premium Spender'

customer_spending['Spender_Category'] = customer_spending['Average_Spending_Score'].apply(categorize_spender)

customer_data = pd.merge(customer_data, customer_spending[['ID', 'Spender_Category']], on='ID', how='left')

customer_data.head()
```

```python
spender_counts = customer_data['Spender_Category'].value_counts()

plt.figure(figsize=(10, 6))
spender_counts.plot(kind='bar', edgecolor='black')
plt.title('Number of Customers by Spending Category')
plt.xlabel('Spender Category')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
purchase_mean = customer_data['PurchaseCount'].mean()
purchase_std = customer_data['PurchaseCount'].std()
```

```python
def categorize_frequency(count):
    if count <= purchase_mean - purchase_std:
        return 'Infrequent Spender'
    elif count <= purchase_mean + purchase_std:
        return 'Moderate Frequency Spender'
    else:
        return 'Frequent Spender'
```

```python
customer_data['Frequency_Category'] = customer_data['PurchaseCount'].apply(categorize_frequency)

customer_data['Combined_Category'] = customer_data['Frequency_Category'] + ' ' + customer_data['Spender_Category']

combined_counts = customer_data['Combined_Category'].value_counts()
```

```python
plt.figure(figsize=(14, 8))
combined_counts.plot(kind='bar', edgecolor='black')
plt.title('Number of Customers by Combined Spending and Frequency Category')
plt.xlabel('Combined Category')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
